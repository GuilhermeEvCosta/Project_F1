**Criando tabela Mediana**

```python
# Calcular a mediana de Starting Grid por pista (Track)
mediana_grid = df_race_consolidado.groupby('Track')['Starting Grid'].median().reset_index()
mediana_grid.columns = ['Track', 'Median Grid']

# Calcular a mediana de Position por pista (Track)
mediana_posicao = df_race_consolidado.groupby('Track')['Position'].median().reset_index()
mediana_posicao.columns = ['Track', 'Median Position']

# Arredondar as medianas para cima
mediana_grid['Median Grid'] = np.ceil(mediana_grid['Median Grid']).astype(int)
mediana_posicao['Median Position'] = np.ceil(mediana_posicao['Median Position']).astype(int)

# Juntar as duas medianas ao DataFrame original
df_race_consolidado_graph = df_race_consolidado.merge(mediana_posicao, on='Track', how='left')
df_race_consolidado_graph = df_race_consolidado_graph.merge(mediana_grid, on='Track', how='left')

# Criar novo DataFrame com as colunas desejadas, sem duplicidade por Track
df_race_median = df_race_consolidado_graph[['Driver', 'Track', 'Median Grid', 'Median Position', 'Team']].drop_duplicates(subset='Track')

df_race_median.head(2)
```

```python
# Salvar o DataFrame Lecrerc Record diretamente no S3
df_race_median.to_csv("s3://testeguilherme-bucket1/f1consolidado/df_race_median.csv", index=False)
```

**Criando tabela Estatistica**

```python
# 1. Criar o DataFrame com os dados da imagem
data_turns = {
    'Track': ['Bahrain', 'Saudi Arabia', 'Australia', 'Emilia Romagna', 'Miami', 'Spain', 'Monaco', 'Azerbaijan', 'Canada', 'Great Britain',
              'Austria', 'France', 'Hungary', 'Belgium', 'Netherlands', 'Italy', 'Singapore', 'Japan', 'United States', 'Mexico',
              'Brazil', 'Abu Dhabi', 'Qatar', 'Las Vegas', 'China'],
    'Turns N¬∞': [57, 50, 58, 63, 57, 66, 78, 51, 70, 52, 71, 53, 70, 44, 72, 53, 61, 27, 56, 71, 71, 58, 57, 50, 56],
    'Turns': ['Long', 'Long', 'Long', 'Short', 'Long', 'Short', 'Short', 'Long', 'Short', 'Long',
              'Short', 'Long', 'Short', 'Long', 'Short', 'Long', 'Short', 'Long', 'Long', 'Short',
              'Short', 'Long', 'Long', 'Long', 'Long']
}

df_turns = pd.DataFrame(data_turns)

# 2. Remover colunas duplicadas de df_race_median, se existirem
df_race_median = df_race_median.drop(columns=['Turns N¬∞', 'Turns', 'Exchange Positions'], errors='ignore')

# 3. Merge limpo
df_race_statics = df_race_median.merge(df_turns, on='Track', how='left')

# 4. Criar coluna 'Exchange Positions' com arredondamento para cima, mantendo sinais negativos
def calcular_exchange(grid, pos):
    diferenca = grid - pos
    if diferenca >= 0:
        return int(np.ceil(diferenca))
    else:
        return int(-np.ceil(abs(diferenca)))

df_race_statics['Exchange Positions'] = df_race_statics.apply(
    lambda row: calcular_exchange(row['Median Grid'], row['Median Position']), axis=1
)

# 5. Fun√ß√£o auxiliar para as colunas Top 3 e out Top 3
def ajustar_valor(median_grid, exchange_position, top3=True):
    if top3 and median_grid in [1, 2, 3]:
        return "-" if exchange_position == 0 else exchange_position
    elif not top3 and median_grid >= 4:
        return "-" if exchange_position == 0 else exchange_position
    else:
        return "-"

# 6. Criar colunas com a l√≥gica de Top 3
df_race_statics['Starting Top 3¬∞'] = df_race_statics.apply(
    lambda row: ajustar_valor(row['Median Grid'], row['Exchange Positions'], top3=True), axis=1
)

df_race_statics['Starting out Top 3¬∞'] = df_race_statics.apply(
    lambda row: ajustar_valor(row['Median Grid'], row['Exchange Positions'], top3=False), axis=1
)

# 7. Salvar resultado final
df_race_statics_new = df_race_statics

# 8. Visualizar os primeiros registros
df_race_statics.head(2)
```

```python
from matplotlib.patches import Patch

# Exemplo de leitura do DataFrame
# df_race_consolidado = pd.read_csv('caminho/para/seuarquivo.csv')

# Remove dados inv√°lidos da coluna 'Position'
df_filtrado = df_race_consolidado[
    ~df_race_consolidado['Position'].isin(['0', 'DQ', 'NC'])
].copy()

# Converte colunas para num√©rico
df_filtrado['Position'] = pd.to_numeric(df_filtrado['Position'], errors='coerce')
df_filtrado['Starting Grid'] = pd.to_numeric(df_filtrado['Starting Grid'], errors='coerce')

# Remove linhas com valores nulos
df_filtrado.dropna(subset=['Starting Grid', 'Position'], inplace=True)

# Conta frequ√™ncia de cada par (Starting Grid, Position)
counts = df_filtrado.groupby(['Starting Grid', 'Position']).size().reset_index(name='count')

# Define cores conforme a frequ√™ncia
def get_color(count):
    if count == 1:
        return 'red'
    elif count == 2:
        return 'yellow'
    elif count == 3:
        return 'gray'
    elif count == 4:
        return 'blue'
    else:
        return 'black'  # opcional para casos acima de 4

colors = [get_color(c) for c in counts['count']]

plt.figure(figsize=(12, 6))

# Cria√ß√£o do gr√°fico
plt.scatter(
    counts['Starting Grid'],
    counts['Position'],
    s=80,
    alpha=0.8,
    c=colors,
    edgecolors='white'
)

# Inverte o eixo Y (1 no topo)
plt.gca().invert_yaxis()

# Define limites manuais para o Y a partir de 1
y_min = 1
y_max = int(counts['Position'].max())
plt.ylim(y_max + 0.5, y_min - 0.5)

# Eixos X e Y com ticks inteiros
x_min = int(counts['Starting Grid'].min())
x_max = int(counts['Starting Grid'].max())
plt.xticks(range(x_min, x_max + 1))
plt.yticks(range(y_min, y_max + 1))

# Labels and Title
plt.xlabel('Starting Position', fontsize=12)
plt.ylabel('Finishing Position', fontsize=12)
plt.title('Performance: Start vs. Finish', fontsize=14)


# Adiciona legenda
legend_elements = [
    Patch(facecolor='red', edgecolor='white', label='Once only'),
    Patch(facecolor='yellow', edgecolor='white', label='Twice'),
    Patch(facecolor='gray', edgecolor='white', label='Three times'),
    Patch(facecolor='blue', edgecolor='white', label='Four times'),
    Patch(facecolor='black', edgecolor='white', label='Five >'),
]
plt.legend(handles=legend_elements, title='Frequency', loc='upper right')

# Remove a grade
plt.grid(False)

plt.show()
```
